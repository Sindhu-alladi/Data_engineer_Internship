```python
spark.sql("CREATE CATALOG IF NOT EXISTS healthcare_new")
```

```python
spark.sql("CREATE SCHEMA IF NOT EXISTS healthcare_new.healthcare_db_sql")
```

```python
spark.sql("USE CATALOG healthcare_new")
spark.sql("USE SCHEMA healthcare_db_sql")
```

```python
spark.sql("""
CREATE TABLE IF NOT EXISTS departments (
  department_id INT,
  department_name STRING
)
USING DELTA
""")
```

```python
spark.sql("""
CREATE TABLE IF NOT EXISTS patients (
  patient_id INT,
  first_name STRING,
  last_name STRING,
  dob DATE,
  gender STRING,
  phone STRING,
  email STRING,
  address STRING
) USING DELTA
""")
```

```python
spark.sql("""
CREATE TABLE IF NOT EXISTS doctors (
  doctor_id INT,
  doctor_name STRING,
  phone STRING,
  email STRING,
  department_id INT
) USING DELTA
""")
```

```python
spark.sql("""
CREATE TABLE IF NOT EXISTS appointments (
  appointment_id INT,
  patient_id INT,
  doctor_id INT,
  appointment_date DATE,
  appointment_time STRING,
  status_1 STRING
) USING DELTA
""")
```

```python
spark.sql("""
CREATE TABLE IF NOT EXISTS diagnoses (
  diagnosis_id INT,
  appointment_id INT,
  diagnosis_details STRING,
  diagnosis_date DATE
) USING DELTA
""")
```

```python
spark.sql("""
CREATE TABLE IF NOT EXISTS prescriptions (
  prescription_id INT,
  appointment_id INT,
  prescription_date DATE
) USING DELTA
""")
```

```python
spark.sql("""
CREATE TABLE IF NOT EXISTS medications (
  medication_id INT,
  medication_name STRING,
  dosage STRING,
  instructions STRING
) USING DELTA
""")
```

```python
spark.sql("""
CREATE TABLE IF NOT EXISTS prescription_medications (
  prescription_id INT,
  medication_id INT,
  quantity INT
) USING DELTA
""")
```

```python
spark.sql("""
CREATE TABLE IF NOT EXISTS lab_tests (
  lab_test_id INT,
  appointment_id INT,
  test_name STRING,
  test_date DATE,
  results STRING
) USING DELTA
""")
```

```python
spark.sql("""
CREATE TABLE IF NOT EXISTS billing (
  bill_id INT,
  appointment_id INT,
  amount DECIMAL(10,2),
  payment_status STRING,
  payment_date DATE
) USING DELTA
""")
```

```python
spark.sql("""
CREATE TABLE IF NOT EXISTS insurance (
  insurance_id INT,
  patient_id INT,
  provider_name STRING,
  policy_number STRING,
  coverage_details STRING
) USING DELTA
""")
```
